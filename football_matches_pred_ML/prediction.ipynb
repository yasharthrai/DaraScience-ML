```python
matches["date"]=pd.to_datetime(matches["date"])
```

## Creating predictors for ML

```python
## Venue_Code 0(Away Game) and 1(Home Game)
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches
```

```python
## Creating a unique code for each opponent
matches["opp_code"]=matches["opponent"].astype("category").cat.codes
matches
```

```python
 ## Creating an hour column for the time(only in hour) on which the matches are played
matches["hour"]=matches["time"].str.replace(":.+", "", regex=True).astype("int")
matches
```

```python
## Creating weekcode i.e a unique code for each day of the week
```

```python
matches["day_code"]=matches["date"].dt.dayofweek
matches
```
