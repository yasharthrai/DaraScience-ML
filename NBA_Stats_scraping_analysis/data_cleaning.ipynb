```python
import pandas as pd
```

```python
mvps=pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps=mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

```python
players=pd.read_csv("players.csv")
players
```

```python
del players["Unnamed: 0"]
```

```python
del players["Rk"]
```

```python
players.head()
```

```python
## Removing astirisk from some names and replacing it with nothing

players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
## Removing iterative names

def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row=df[df["Tm"] == "TOT"]
        row["Tm"] =df.iloc[-1,:]["Tm"]
        return row
    
players=players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(50)
```

```python
## Two Times
players.index=players.index.droplevel()
```

```python
players.head(50)
```

```python
combined=players.merge(mvps, how="outer", on=["Player", "Year"])
combined.head()
```

```python
combined[combined["Pts Won"] > 0]
```

```python
## combined[["Pts Won", "Pts Max", "Share"]].fillna(0, inplace=True)

## OR We can Write 

combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0, inplace=True)
```

```python
combined.head()
```

```python
teams=pd.read_csv("teams.csv")
teams
```

```python
teams.head(30)
```

```python
##Deleting the divisions column

teams=teams[~teams["W"].str.contains("Division")]
teams.head(30)
```

```python
## Removing the astirisk* from team names

teams["Team"]=teams["Team"].str.replace("*", "", regex=False)
teams.head(30)
```

```python
pd.read_csv("nicknames.csv")
```

```python
nicknames={}

with open("nicknames.csv") as f:
    lines=f.readlines()
```

```python
lines
```

```python
for line in lines:   ## From 1 onwards
    abbrev,name = line.replace("\n","").split(",")
    nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats=combined.merge(teams, how="outer", on=["Team", "Year"])
stats
```

```python
del stats["Unnamed: 0_x"]
```

```python
del stats["Rk"]
```

```python
del stats["Unnamed: 0_y"]
```

```python
stats
```

```python
stats.dtypes
```

```python
## Changing the data types from object to numeric and ignring (not converting ) the ones that give errors

stats=stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"]=stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python
stats["GB"]=pd.to_numeric(stats["GB"])
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
highest_scoring=stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
highest_scoring
```

```python
highest_scoring.plot.bar("Player", "PTS")
```
