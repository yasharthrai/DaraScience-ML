```python
## Extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
batting_first.shape 
```

```python
## Making a histogram
plt.figure(figsize=(5,5))
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
batting_first['winner']
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=['r','g','b'])
plt.show()
```

```python
## Pie-chart for top 5 teams
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()[0:5]),labels=list(batting_first['winner'].value_counts()[0:5].keys()),autopct='%0.2f%%')
plt.show()
```

```python
## Pie-chart for all the teams
plt.figure(figsize=(17,17))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.2f%%')
plt.show()
```
