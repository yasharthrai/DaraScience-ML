```python
## Extracting individual columns
```

```python
unmarried=census[census['relationship']=='Unmarried']
unmarried.head()
```

```python
divorced=census[census['marital.status']=='Divorced']
divorced.head()
```

```python
divorced.shape
```

```python
old_male=census[(census['age']>50) & (census['sex']=='Male')]
old_male.head()
```

```python
old_male.shape
```

```python
white_income=census[(census['race']=='White') & (census['income']=='>50K') & (census['marital.status']=='Never-married')]
white_income.head()
```

 ### Importing Seaborn as sns
 

```python
import seaborn as sns
from matplotlib import pyplot as plt   ## OR 'import matplotlib.pyplot as plt'
```

```python
## SNS DistPlot for numerical values
```

```python
sns.distplot(census['age'])
plt.show()
```

```python
sns.distplot(census['hours.worked'])
plt.show()
```

#### BarPlot ( Categorical Columns on X-Axis and numerical on Y-Axis)

```python
sns.barplot(x='sex',y='hours.worked',data=census)
plt.show()
```

```python
sns.barplot(x='sex',y='hours.worked',data=census,hue='race')
plt.show()
```

```python
sns.barplot(x='sex',y='hours.worked',data=census,hue='education')
plt.show()
```

```python
sns.barplot(x='sex',y='hours.worked',data=census,hue='marital.status')
plt.show()
```

#### Using Lineplot for comparison b/w two numerical entities

```python
sns.lineplot(x='age',y='hours.worked',data=census)
plt.show()
```

```python
sns.lineplot(x='age',y='hours.worked',data=census,hue='sex')
plt.show()
```

```python
plt.figure(figsize=(25,25))
sns.lineplot(x='age',y='hours.worked',data=census,hue='race')
plt.show()
```
