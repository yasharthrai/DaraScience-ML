## Data Cleaning
- year has many non-year numbers
- year object to int
- --
- price has ask for price
- price object to int
- --
- kms_driven has kms with integers
- kms_driven object top int
- --
- fuel_type has NaN values
- --
- keep first three words only from the name column

```python
 backup=car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

```python
car=car[car['Price']!="Ask for Price"]
```

```python
car['Price'] = car['Price'].str.replace(',', '', regex=True)
car['Price'] = pd.to_numeric(car['Price'], errors='coerce')
```

```python
car = car.dropna(subset=['Price']).astype({'Price': int})
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','', regex=True)
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car.info()
```

```python
car=car[~car['fuel_type'].isna()]
```

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ') #3 is excluded means index 0 to 2
```

```python
car
```

```python
car=car.reset_index(drop=True) ## to reset the index and drop the previous index and 
```

```python
car.describe()
```

```python
car[car['Price']>6e6]
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

# Model

```python
x=car.drop(columns='Price')
y=car["Price"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_preds=pipe.predict(x_test)
```

```python
y_preds
```

```python
r2_score(y_test, y_preds)
```

```python
## running the above processes several times for a large dataset
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_preds=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_preds))
    
```

```python
scores
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
## Now training the model at the random state that gives the maximum r2_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_preds=pipe.predict(x_test)
r2_score(y_test,y_preds)
```

```python
## Now dumping(saving) the pipeline using pickle
import pickle
pickle.dump(pipe,open('Linearregression.pkl','wb'))
```
